```python
#The following analysis was created using February 2019 bike trip data from: https://www.kaggle.com/code/nancyalaswad90/data-visualization-trip-data-ford-go-bike/data
```

```python
import pandas as pd
```

```python
#import and look at the data
bikedata=pd.read_csv('/content/fordgobike-tripdataFor201902.csv')
bikedata.head(10)
```

```python
#create and calculate columns for minutes and hours of duration
bikedata['duration_min']=bikedata['duration_sec']/60
bikedata['duration_hrs']=bikedata['duration_min']/60
bikedata.head(10)
```

```python
#Has any particular bike been used more than its peers (i.e. should the bikeshare operater consider special maintenance schedules or base maintenance on time such as every two weeks, every month, etc?)
#Select id and hour duration; drop na
bikeuse=bikedata[['bike_id','duration_hrs']]
bikeuse.dropna()
#Group data by bike id and sum the hrs ridden; sort desc
bikeuse=bikeuse.groupby('bike_id').sum().sort_values('duration_hrs',ascending=False)
bikeuse.head(50)
```

```python
#Visualize the data by plotting it on a histogram with duration as the x axis

import matplotlib.pyplot as plt
import statistics
import numpy as np

#Assign axis; calculate mean, standard dev, and quartiles for analysis
x=bikeuse['duration_hrs']
mean=statistics.mean(x)
std=statistics.stdev(x)
firstq=np.percentile(x,q=25)
secondq=np.percentile(x,q=50)
thirdq=np.percentile(x,q=75)
upperpercentile=np.percentile(x,q=90)

plt.hist(x,bins=50)
plt.title("Hours Used Histogram")
plt.xlabel("Hours Used")
plt.ylabel("Number of Bikes")

#Add lines on the histogram for mean and one standard dev
plt.axvline(mean,color='k',linestyle='dashed')
plt.axvline(mean+std,color='y',linestyle='dashed')
plt.axvline(mean-std,color='y',linestyle='dashed')

#Show the plot and print the mean, standard deveation, and quartiles + 90th percentile
plt.show()
print("Mean = " + str(mean)) 
print("Standard Deviation = " + str(std))
print("First Quartile = " + str(firstq))
print("Median = " + str(secondq))
print("Third Quartile = " + str(thirdq))
print("90th Percentile = " + str(upperpercentile))
```

```python
#The histogram shows that half of the bikes were used for 5 hours. 90% of all bikes were used less than about 21 hours.
#The operator might want to identify bikes used more frequently to either accelerate its meintenance schedule or prioritize use of other bikes.
#Bikes in the 90th percentile and above were rented for about an hour for every business day in the month.
```

```python
#Can a model be constructed to predict ride length?
import sklearn as sk
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

#Select data for the regression - dependent variable duration, and ind. variables user type, birth year, gender
regdata=bikedata[['duration_sec','user_type','member_birth_year','member_gender']]
regdataclean=regdata.dropna()

#Encode categorical variables
#user_type
encoder=LabelEncoder()
userencoded=regdataclean[['user_type']]
userencoded['user_type'] = encoder.fit_transform(regdataclean['user_type'])
userencoded.value_counts()
```

```python
#gender
encoder=LabelEncoder()
genderencoded=regdataclean[['member_gender']]
genderencoded['member_gender'] = encoder.fit_transform(regdataclean['member_gender'])
genderencoded.value_counts()
```

```python
#Add encoded data back to a new df as renamed columns
regdataencoded=regdataclean
regdataencoded['user_type_encoded']=userencoded
regdataencoded['member_gender_encoded']=genderencoded
print(regdataencoded)
```

```python
#calculate age and drop old columns
regdataencoded['age']=2019-regdataencoded['member_birth_year']
regdatafinal=regdataencoded[['duration_sec','user_type_encoded','member_gender_encoded','age']]
regdatafinal.head(50)
```

```python
#Create dfs for dependent and independent variables
y=regdatafinal['duration_sec']
X=regdatafinal[['user_type_encoded','member_gender_encoded','age']]

#Split into training and testing sets with 1/3 as test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.333)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
#Fit the model
from sklearn import linear_model

lm=linear_model.LinearRegression()
model=lm.fit(X_train,y_train)
predictions=lm.predict(X_test)
predictions
```

```python
#Plot the values vs predictions
plt.scatter(y_test, predictions)
plt.xlabel('True Values')
plt.ylabel('Predictions')
```

```python
print('Score:', model.score(X_test, y_test))
```

```python
#Cross validate the model
from sklearn.model_selection import cross_val_score, cross_val_predict
from sklearn import metrics

scores=cross_val_score(model,X,y,cv=5)
print('Cross-validated scores:', scores)
```

```python
predictions=cross_val_predict(model,X,y,cv=5)
plt.scatter(y,predictions)
```

```python
accuracy=metrics.r2_score(y,predictions)
print('Cross-Predicted Accuracy:', accuracy)
```

```python
#The simple model and cross validated model indicate that we cannot reliably predict a ride durations based off of
#user data including gender, user type, or age.
```
